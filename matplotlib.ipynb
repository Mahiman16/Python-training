```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##line clot

plt.plot(arr)
# plt.sort()
plt.show()
# plt.sort()
```

```python
x = [1,2,3,4,5]
y = [1,10,9,30,25]
plt.plot(x,y,marker = "d",ms = 10,
         mfc = "red",mec = "purple",color = "black",
         lw = 5,ls = ":",
         alpha = 0.6,label = "lineplot")
plt.title("lineplot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

```python
##scatterplot
# plt.scatter(x,y)
plt.scatter(x,y,marker = "d")
plt.show()
```

```python
x=["mp","up","rj","gj"]
y=[10,20,30,40]
colours = ["red","green","blue","pink"]
hatch_patterns= ["//","\\\\","xx","oo"]
plt.bar(x,y,color = colours,hatch = hatch_patterns,label = x,width = 0.3,edgecolor = "yellow",alpha = 0.7,
        linewidth = 3)
plt.legend()
plt.show()
```

```python
x=["mp","up","rj","gj"]
y=[10,20,30,40]
colours = ["red","green","blue","pink"]
hatch_patterns= ["//","\\\\","xx","oo"]
# plt.barh(x,y,color = colours,hatch = hatch_patterns,label = x,width = 0.3,edgecolor = "yellow",alpha = 0.7,
        # linewidth = 3)
plt.barh(x,y)
plt.legend()
plt.grid()
plt.show()
```

```python
##histogram -- distribution of data in intervals(bins)
arr = np.random.randint(20000,40000,100)
plt.hist(arr,bins=20)
plt.show()
```

```python
x=["mp","up","rj","gj"]
y=[10,20,30,40]
colours = ["red","green","blue","pink"]
explodes = (0.4,0.2,0.9,0)
plt.pie(y,labels = x,autopct="%1.2f%%",shadow = True,explode=explodes)
##save chart
plt.savefig("piechart.pdf")
plt.show()
```

```python
##subpplot
x = np.linspace(0,10,100)
plt.figure(figsize=(10,6))
plt.subplot(2,2,1)#2 rows #columns,1st graph
plt.plot(x,np.sin(x))
plt.title("sin")
plt.subplot(2,2,2)#2 rows #columns,1st graph
plt.plot(x,np.cos(x))
plt.title("cos")
plt.subplot(2,2,3)#2 rows #columns,1st graph
plt.plot(x,np.tan(x))
plt.title("tan")
plt.subplot(2,2,4)#2 rows #columns,1st graph
plt.plot(x,np.log(x))
plt.title("log")

```
